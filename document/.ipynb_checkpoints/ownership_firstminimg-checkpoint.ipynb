# <h1>1.股權分散圖資料初步探索</h2>

```python
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine
import seaborn as sns

#建立與 MySQL 的連結
class mysql_engine():
 user='root'
 passwd='ian1991'
 host='localhost'
 port = '3306'
 db_name='tfb103d_project'
 engine = create_engine('mysql+pymysql://{0}:{1}@{2}:{3}/{4}?charset=utf8'.format(user,passwd,host,port,db_name))

#設定讀取 MySQL 中的資料表並轉為 DataFrame 的功能
def get_data(sql):
 pg_enine=mysql_engine()
 try:
  with pg_enine.engine.connect() as con, con.begin():
   df=pd.read_sql(sql,con)# 獲取資料
  con.close()
 except:
  df=None
 return df
```

以 get_data 讀取整張股權分散圖 (ownership) 的資料表，包含特徵如下:
>資料日期, data_date        
>集保總張數, total_board      
>總股東人數, total_holder     
>平均張數/人, avg_board        
>\>400張大股東持有張數, over400_amount   
>\>400張大股東持有百分比, over400_rate     
>\>400張大股東人數, over400_holder   
400-600張人數, between400_600   
600-800張人數, between600_800   
800-1000張人數, between800_1000  
>1000張人數, over1000_holder  
>1000張大股東持有百分比, over1000_rate    
收盤價, closing_price    

```python
ownership = get_data("""SELECT * FROM tfb103d_project.ownership;""")
ownership.head(10)
```

<h2>先看一下股東們的分布趨勢 (2021-09-24)

```python
holder = get_data("""SELECT stock_code,
                            ((total_holder-over400_holder)/total_holder) under400_rate, 
                            between400_600/total_holder between400_600_rate, 
                            between600_800/total_holder between600_800_rate, 
                            between800_1000/total_holder between800_1000_rate,
                            over1000_holder/total_holder over1000_rate,
                            over400_holder/total_holder over400_rate
                    FROM tfb103d_project.ownership
                    WHERE data_date = '2021-09-24';""")

holder
```

```python
fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(2,2,1) # row, colimn, 位置
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

data1 = list(holder.iloc[0,2:])
ax1.bar(['400_600','600_800','800_1000','>1000','>400'],data1)
ax1.set_title('2330')

data2 = list(holder.iloc[1,2:])
ax2.bar(['400_600','600_800','800_1000','>1000','>400'],data2)
ax2.set_title('2303')

data3 = list(holder.iloc[2,2:])
ax3.bar(['400_600','600_800','800_1000','>1000','>400'],data3)
ax3.set_title('2379')

data4 = list(holder.iloc[3,2:])
ax4.bar(['400_600','600_800','800_1000','>1000','>400'],data4)
ax4.set_title('6488')

plt.show()
```

<h3> 400張的股東分布趨勢看起來各公司都蠻相近的，落差不大

```python
holder_all = get_data("""SELECT stock_code,
                            ((total_holder-over400_holder)/total_holder) under400_rate, 
                            between400_600/total_holder between400_600_rate, 
                            between600_800/total_holder between600_800_rate, 
                            between800_1000/total_holder between800_1000_rate,
                            over1000_holder/total_holder over1000_rate,
                            over400_holder/total_holder over400_rate
                    FROM tfb103d_project.ownership;""")

holder_all.head(10)
```

```python
corr = holder_all.corr()
sns.heatmap(corr,cmap="RdBu_r")
```

```python
sns.pairplot(holder_all)
```

<h3>綜上討論，大於400張股東的分布變化趨勢很接近，可以取一項代表即可；另小於400張的股東可做為小股東(散戶)的代表

來看一下持股比例跟股價的關係

```python
ownership = get_data("SELECT * FROM tfb103d_project.ownership Where stock_code = 2303 order by data_date;")
# print(ownership)
# ownership.info()
over400_rate = ownership['over400_rate']
over1000_rate = ownership['over1000_rate']
price = ownership['closing_price']
date = ownership["data_date"]
```

```python
fig = plt.figure(figsize=(12,3))
ax1 = fig.add_subplot(1,2,1) # row, colimn, 位置
ax2 = fig.add_subplot(1,2,2)

ax1.scatter(price,over400_rate) 
ax2.scatter(price,over1000_rate) 
```

<h3>大股東持股數似乎跟股價有負相關性

```python
under400_rate = []
for i in over400_rate:
    under400_rate.append(round(100-float(i),2))
# print(under400_rate)
```

```python
ratio = over1000_rate/under400_rate
plt.scatter(price,ratio) 
# plt.scatter(price,over1000_rate) 
plt.xlabel('price') 
# plt.ylabel('under400_rate') 
plt.title('2330')
```

<h3>比較大小股東的持股比率跟價錢的關係，也觀察到股價與大小股東的持股比率似乎有負相關性

# 2.董監事持股資料

```python
DS_ownership = get_data("SELECT * FROM tfb103d_project.dsownership where data_date != '2021-10-01' and data_date != '2021-09-01';")
print(DS_ownership)
```

```python
ds_2303 = DS_ownership.loc[DS_ownership.stock_code == "2303"]
ds_2330 = DS_ownership.loc[DS_ownership.stock_code == "2330"]
ds_2379 = DS_ownership.loc[DS_ownership.stock_code == "2379"]
ds_6488 = DS_ownership.loc[DS_ownership.stock_code == "6488"]
# ds_6488.head(10)
```

```python
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(2,2,1) # row, colimn, 位置
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(ds_2303['data_date'],ds_2303['toldirector_rate'],"red") 
ax1.plot(ds_2303['data_date'],ds_2303['month_price']/8,"blue") 

ax2.plot(ds_2330['data_date'],ds_2330['toldirector_rate'],"red") 
ax2.plot(ds_2330['data_date'],ds_2330['month_price']/50,"blue") 

ax3.plot(ds_2379['data_date'],ds_2379['toldirector_rate'],"red") 
ax3.plot(ds_2379['data_date'],ds_2379['month_price']/50,"blue") 

ax4.plot(ds_6488['data_date'],ds_6488['toldirector_rate'],"red") 
ax4.plot(ds_6488['data_date'],ds_6488['month_price']/10,"blue") 
```

```python
DS_ownership_price = get_data("SELECT stock_code, data_date, month_price,rate_of_change, director_rate, indirector_rate,toldirector_rate,foreign_rate FROM tfb103d_project.dsownership where data_date != '2021-10-01' and data_date != '2021-09-01';")
DS_ownership_price_1 = DS_ownership_price.drop("stock_code",axis=1)
DS_ownership_price_1 = DS_ownership_price_1.drop("data_date",axis=1)
```

```python
# ownership_h['data_date'] = ownership_h['data_date'].astype(float)
corr = DS_ownership_price_1.corr()
sns.heatmap(corr,cmap="RdBu_r")
```

```python
sns.pairplot(DS_ownership_price_1)
```

<h3>整體來說，各筆資料間無明顯相關性；其中，外資與股東持股雖呈現負相關，但據散布圖的趨勢看不出有分布有連續性，因此先保守估計無相關性。

<h3> 全體董監事持股比例與獨/非獨董監事趨勢相當，可具代表性。(toldirector_rate)
外資持股(foreign__rate) 與漲跌幅(rate_of_change) 初步看不出有明顯的相關性，可再進一步分析。
